```python
import numpy as np

import pyvista as pv
```

```python
# Make data
x = np.arange(-10, 10, 0.25)
y = np.arange(-10, 10, 0.25)
x, y = np.meshgrid(x, y)
r = np.sqrt(x**2 + y**2)
z = np.sin(r)
```

```python
# Create and plot structured grid
grid = pv.StructuredGrid(x, y, z)
# Plot mean curvature as well
grid.plot_curvature(clim=[-1, 1])
```
